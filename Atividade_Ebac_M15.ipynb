# 3) Identifique e trate as colunas que contém outliers.
Caso opte por mante-los ou altera-los justifique sua escolha.


```python
# ---------- Verificando erros/outliers. ----------
df['Genero'].unique()
```

```python
# ---------- Tratando a coluna 'Genero'. -----------
df['Genero'].unique()

df['Genero'] = df['Genero'].str.capitalize()

#Tratando os valores.
df['Genero'] = df['Genero'].replace({
    'f' : 'Female', 'F' : 'Female',
    'm' : 'Male', 'M' : 'Male',})
```

```python
# ---------- PLotando o gráfico. ----------
plt.figure(figsize=(6,6))
sns.boxplot(x=df['Genero'], widths=0.5)
plt.title('Box Plot de Genero')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

* R: Fiz o tratamento aqui porque poderiam ser erros de digitação transformando 'f', 'F' em 'Female' e 'm', 'M' em 'Male' tratando os outliers da melhor forma.

```python
# ---------- Tratando os outliers da variável "Tempo_como_Cliente. ----------
Q1 = df['Tempo_como_Cliente'].quantile(0.25)
Q3 = df['Tempo_como_Cliente'].quantile(0.75)
IQR = Q3 - Q1

limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

df_sem_outliers = df[(df['Tempo_como_Cliente'] >= limite_inferior) & (df['Tempo_como_Cliente'] <= limite_superior)]
print(df_sem_outliers['Tempo_como_Cliente'].describe())
```

```python
# ---------- Plotando o gráfico para conferir se os outliers foram tratados. ----------
plt.figure(figsize=(6,3))
sns.boxplot(df, x='Tempo_como_Cliente')
plt.title('Box Plot de Tempo_como_Cliente')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# ---------- Gráfico de barras empilhadas Churn x Tipo_Contrato. ----------
df_grouped = (df.groupby(['Churn', 'Tipo_Contrato']).size().reset_index(name='Count'))
df_grouped['Total_Churn'] = df_grouped.groupby('Churn')['Count'].transform('sum')
df_grouped['Percent'] = (df_grouped['Count'] / df_grouped['Total_Churn']) * 100

# ----------- Gráfico de barras empilhadas. ----------
# .round(2) ele traz casas decimais.
fig = px.bar(df_grouped, x='Churn', y='Percent',  color='Tipo_Contrato', barmode='stack', text=df_grouped['Percent'].round(2).astype(str) + '%',labels={
        'Churn': 'Churn',
        'Percent': 'Porcentagem',
        'Tipo_Contrato': 'Tipo de Contrato'},
    title='Distribuição Percentual do Tipo de Contrato por Churn')
fig.show()
```

### Trazendo insigths
* Nesse gráfico Observamos que 89,68% dos clientes que fazem contratos mensalmente, são mais propensos a dar chrun. Ou melhor, abandonar.
* Clientes com 1 ano de Contrato tem 8,94% de chances de churn, um valor bem baixo, e melhor do que mensalmente.
* E por fim, cliente que tem 2 anos de contratos tem 1,39% de chances de dar Churn.

```python
# ---------- Gráfico de barras empilhadas Churn x Idoso. ----------
df_grouped = (df.groupby(['Churn', 'Idoso']).size().reset_index(name='Count'))
df_grouped['Idoso'] = df_grouped.groupby('Churn')['Count'].transform('sum')
df_grouped['Percent'] = (df_grouped['Count'] / df_grouped['Idoso']) * 100

# ----------- Gráfico de barras empilhadas. ----------
# .round(2) ele traz casas decimais.
fig = px.bar(df_grouped, x='Churn', y='Percent',  color='Idoso', barmode='stack', text=df_grouped['Percent'].round(2).astype(str) + '%',labels={
        'Churn': 'Churn',
        'Percent': 'Porcentagem',
        'Idoso': 'Idoso'},
    title='Distribuição Percentual Churn por Idoso')
fig.show()
```

```python
# ---------- Gráfico de barras empilhadas Churn x FormaPagamento. ----------
df_grouped = (df.groupby(['Churn', 'FormaPagamento']).size().reset_index(name='Count'))
df_grouped['FormaPagamento'] = df_grouped.groupby('Churn')['Count'].transform('sum')
df_grouped['Percent'] = (df_grouped['Count'] / df_grouped['FormaPagamento']) * 100

# ----------- Gráfico de barras empilhadas. ----------
# .round(2) ele traz casas decimais.
fig = px.bar(df_grouped, x='Churn', y='Percent',  color='FormaPagamento', barmode='stack', text=df_grouped['Percent'].round(2).astype(str) + '%',labels={
        'Churn': 'Churn',
        'Percent': 'Porcentagem',
        'FormaPagamento': 'Forma de Pagamento'},
    title='Distribuição Percentual da Forma de Pagamento por Churn')
fig.show()
```

# Quais as variáveis mais importante relacionadas a Churn.

* Tipo_Contrato: Ela nos mostra que, clientes mensais tendem a ter maiores chances de Churn. Enquanto o contrato anul e de anos, tem uma tendência menor em Churn.

* Tempo_Como_Cliente: Ela nos mostra o tempo em que o ficou e quando deu churn.

* FormaPagamento: Vimos que, Cliente que usam "Eletric Check" tendem a dar mais churn, do que os cliente que usam pagamento "Automáticos". Habilitando novas formas pagamento, levando o cliente a ter mais "comodidade" poderiamos ter uma que grande nas pessoas que dão churn.

* Idoso: Ele dão mais churn por diversos motivos. Por falta de conhecimento com a tecnologia, por não saber utilizarem os metodos mais fáceis. Aos cliente que são idoso deveriamos dar uma atenção diferente, para que o chrun pudesse ser reduzido.
